```python
import pandas as pd
import random
```

```python
data = pd.read_csv('../reste_des_formations.csv')
data['LIENS'] = 'https://www.coursera.org' + data['LIENS'] 
```

```python
formateurs_id = {

}
for f,i in zip(data.FORMATEUR.unique(),range(997,1144)):
    formateurs_id[f] = i
    print("INSERT INTO FORMATEURS (NOM_FORMATEUR) VALUES ('"+f+"');")
```

```python
formateurs_id
```

```python
start = 100000000
end = 799999999
```

```python
data['id_formateur'] = [formateurs_id.get(formateurr) for formateurr in data.FORMATEUR]
data['Telephone'] = None
```

```python
data['adresse'] = "en ligne"
data['Telephone'] =  ["0" + str(random.randint(start, end)) for f in data.FORMATEUR]
data.Telephone = [' '.join(numero[i:i+2] for i in range(0, len(numero), 2)) for numero in data['Telephone']]
```

```python
newdata = data[['ID_FORMATION','id_formateur','Telephone','adresse','LIENS']]
```

```python
newdata.to_csv('formateurs_reste_pret.csv.csv', index=False)
```
